## Exploratory data analysis

### Life expectancy across the years

```python
plt.figure()
sns.lineplot(data=df, x='Year',y='Life', ci=None, hue='Country')
plt.show()
```

Most of the country considered have a life expectancy at birth above 70 years, except for Zimbawe. However Zimbawe is a developing country, and we can see that since 2004 its life expectancy is growing.

Up to 2004 there was the problem of people dying of hunger. All the other countries have their life expectancy growing as well

```python
plt.figure()
sns.scatterplot(data=df, x='GDP',y='Life', ci=None, hue='Country')
plt.show()
```

The above chart does not allow to find a clear relation between Life and GDP due to the outlier Zimbawe. 

Actually we can see that when GDP increases, also life expectance increases, but let's take a closer look.

```python
plt.figure()
sns.scatterplot(data=df, x='GDP',y='Life', ci=None, hue='Country')
plt.ylim((70, 85))
plt.show()
```

We can see that there is a linear relationship between GDP and life expectancy. China is the only country that has an almost logaritmic life expectancy / GDP correlation

```python
plt.figure(figsize=(20,15))
sns.violinplot(data=df, x='Country',y='Life')
```

Zimbawe has the highest life expecancy span with many outliers. Other countries have less outliers and a minor range.

```python
sns.histplot(df, x='GDP', bins=30)
```

GDP data are skewed to the right. Most of the data considered have a low GDP
